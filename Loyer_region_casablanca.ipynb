```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/montants-loyers-region-casablanca-settat-2001-2019 1.xlsx")
```

```python
df.info()
```

```python
import seaborn as sns

# Préparer les données pour la heatmap
df_pivot = df.pivot_table(index='Agglomération', columns='Type d\'habitat', values=2001, aggfunc='mean')

# Création de la heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(df_pivot, annot=True, cmap='YlGnBu', fmt='.1f')
plt.title('Heatmap des loyers en 2019 par agglomération et type d\'habitat', fontsize=14)
plt.xlabel('Type d\'habitat', fontsize=12)
plt.ylabel('Agglomération', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Supposons que df est votre DataFrame

# Liste des années
years = [year for year in range(2001, 2020)]

# Filtrer les colonnes pour obtenir uniquement les données des années
df_years = df.set_index(['Agglomération', 'Type d\'habitat', 'Sous type']).loc[:, years]

# Créer des heatmaps pour chaque combinaison de type et sous-type
types = df['Type d\'habitat'].unique()
subtypes = df['Sous type'].unique()

for type_habitat in types:
    for subtype in subtypes:
        # Filtrer les données pour le type et sous-type courants
        subset = df[(df['Type d\'habitat'] == type_habitat) & (df['Sous type'] == subtype)]
        df_pivot = subset.set_index('Agglomération').loc[:, years]

        # Créer la heatmap
        plt.figure(figsize=(20, 10))
        sns.heatmap(df_pivot, annot=True, cmap='YlGnBu', fmt='.1f', linewidths=.5)
        plt.title(f'Heatmap des loyers pour Type: {type_habitat} et Sous-type: {subtype}', fontsize=14)
        plt.xlabel('Année', fontsize=12)
        plt.ylabel('Agglomération', fontsize=12)
        plt.xticks(rotation=45)
        plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the trend for each "Agglomération" over time
plt.figure(figsize=(12, 8))
for agglomeration in df['Agglomération'].unique():
    agglomeration_data = df[(df['Agglomération'] == agglomeration) & (df["Type d'habitat"] == 'Appartement économique')].loc[:, 2001:2019].mean()
    plt.plot(agglomeration_data, label=agglomeration)

# Customize the x-axis ticks to show integer years
plt.xticks(ticks=range(2001, 2020), labels=range(2001, 2020))  # Ensure integer years on x-axis

plt.title("Loyer Evolution (economical apartment) by Agglomération (2001-2019)")
plt.xlabel("Year")
plt.ylabel("Average Loyer")
plt.legend()
plt.grid(True)  # Optional: Add a grid for better readability
plt.show()
```

```python
# Boxplot to compare rent values for different habitat types
plt.figure(figsize=(12, 8))
df_melted = df.melt(id_vars=['Agglomération', 'Type d\'habitat'], value_vars=[year for year in range(2001, 2020)], var_name='Year', value_name='Rent')

sns.boxplot(x='Type d\'habitat', y='Rent', data=df_melted)
plt.title("Rent Comparison by Type of Habitat (2001-2019)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the percentage change in rent values year over year for each "Agglomération"
df_change = df_casablanca.set_index('Agglomération').loc[:, 2001:2019].pct_change(axis=1)

# Visualize the rate of change
plt.figure(figsize=(12, 8))
sns.heatmap(df_change, annot=True, cmap='coolwarm', fmt='.2%', linewidths=.5)
plt.title('Year-over-Year Rent Growth (%) by Agglomération (2001-2019)')
plt.xlabel('Year')
plt.ylabel('Agglomération')
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = df.set_index('Agglomération').loc[:, 2001:2019].corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Between Years')
plt.show()
```
